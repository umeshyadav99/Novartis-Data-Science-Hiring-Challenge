# Loading training dataset

```python
train = pd.read_csv('/kaggle/input/novartis/Train.csv')
train.head(5)
```

# Importing python datetime library

```python
import datetime
#train['month'] = pd.DatetimeIndex(train['DATE']).month
train['year']  = pd.DatetimeIndex(train['DATE']).year
```

```python
train['MULTIPLE_OFFENSE1']=train['MULTIPLE_OFFENSE']
```

```python
train = train.drop(['MULTIPLE_OFFENSE'],axis=1)
```

```python
train.head(4)
```

```python
X = train.iloc[: ,2:-1].values
y = train.iloc[: , -1].values
X
```

# Applying SimpleImputer method of sklearn library to handle missing value

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
imputer = imputer.fit(X)
X = imputer.transform(X)
```

```python
y
```

```python
#from sklearn.model_selection import train_test_split
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

# Applying StandardScaler method of sklearn.preprocessing library to apply feature scaling

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X = sc_X.fit_transform(X)
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(random_state=1,
                             n_estimators=190,
                             criterion='gini',
                            max_depth=25,
                             min_samples_split=2,
                             min_samples_leaf=1,
                             bootstrap=True,
                            n_jobs=-1,verbose=False)

clf.fit(X,y)
```

# XGB classifier from xgboost

```python
from xgboost import XGBClassifier
classifier_xgb = XGBClassifier()
classifier_xgb.fit(X,y)
```

```python
#clf.score(X_test,y_test)
```

# Loading test data

```python
test = pd.read_csv('/kaggle/input/novartis/Test.csv')
test.head()
```

```python
import datetime
#test['month'] = pd.DatetimeIndex(test['DATE']).month
test['year']  = pd.DatetimeIndex(test['DATE']).year
```

```python
test1 = test.iloc[:,2:]
test1.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
test1 = sc_X.fit_transform(test1)
```

```python
prediction = classifier_xgb.predict(test1)
prediction
```
